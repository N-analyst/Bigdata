## !! 읽어온데이터 수정
## !! 공릉1.3동->공릉1동
## !! 답십리제3동 -> 답십리1동

```python
dong_sum=data.groupby(['동/읍/면']).sum()
print(dong_sum.loc['답십리제3동'])
print(dong_sum.loc['공릉1.3동'])
dong_sum.rename(index={'답십리제3동':'답십리1동','공릉1.3동':'공릉1동'},inplace=True)
city=dong_sum.index
city
```

```python
city=list(map(lambda x: x.replace('.','·'),city))
import re
p=re.compile('제\d+')
for i in range(len(city)):
    if p.search(city[i]):
        start=p.search(city[i]).start()
        city[i]=city[i][:start]+city[i][start+1:]
print(city)
print(len(city))
```

```python
# map.json파일에 있는 모든 동 이름 리스트
geomap_namelist=[]
for dong in geo_data['features']:
    dong_name=dong['properties']['name']
    geomap_namelist.append(dong_name)
```

```python
err_list=[]
for dong in city:
    if not dong in geomap_namelist:
        err_list.append(dong)
print(err_list)
print(len(err_list))
```

```python
dong_sum.index=city
dong_sum.loc[dong_sum.index.isin(err_list)]
```

```python
new_geo_data={'type':'FeatureCollection','features':[]}
name_list=[]
for dong in geo_data['features']:
    dong_name=dong['properties']['name']
    if dong_name in city:
        new_geo_data['features'].append(dong)
        name_list.append(dong_name)        
new_geo_data
```

```python
# 팝업처리 추가
import pandas as pd
import folium
import json

# 경기도 인구변화 데이터를 불러와서 데이터프레임으로 변환
file_path = './data/total_top100.xlsx'
df = pd.read_excel(file_path)
df=df.groupby(['동/읍/면']).sum()
df.index=city
df=df.dropna(axis=0)
#df.columns = df.columns.map(str)
display(df)
# 경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기
geo_path = './data/new_map.json'
try:
    geo_data = json.load(open(geo_path, encoding='cp949'))
except:
    geo_data = json.load(open(geo_path, encoding='utf-8-sig'))

# 서울 지도 만들기
g_map = folium.Map(location=[37.566345, 126.977893], 
                   tiles='Stamen Terrain', zoom_start=9)



# Choropleth 클래스로 단계구분도 표시하기
aa=folium.Choropleth(geo_data=geo_data,    # 지도 경계
                 data = df['승차건수'],      # 표시하려는 데이터
                 columns = [df.index, df['승차건수']],  # 열 지정
                 fill_color='YlOrRd', fill_opacity=0.7,
                 nan_fill_opacity=0, smooth_factor=0.5,bins=9,             
                 key_on='feature.properties.name').add_to(g_map)
aa.geojson.add_child(
    folium.features.GeoJsonTooltip(['name'],labels=False)
    #folium.features.GeoJsonDetail(['name'],labels=True)
)



for i in charger_clean.index:
    lat = charger_clean.loc[i,'위도']
    lng = charger_clean.loc[i,'경도']
    
    #추출한 정보를 지도에 표시
    icon = folium.Icon(color='orange', icon='flash')
    tooltip=charger_clean.loc[i,'소재지도로명주소']
    marker = folium.Marker([lat,lng],icon=icon,tooltip=tooltip)
    marker.add_to(g_map)

for i in library.index:
    lat = library.loc[i,'위도']
    lng = library.loc[i,'경도']
    
    #추출한 정보를 지도에 표시
    icon = folium.Icon(color='blue', icon='book')
    #tooltip=charger_clean.loc[i,'소재지도로명주소']
    marker = folium.Marker([lat,lng],icon=icon)
    marker.add_to(g_map)
    
for i in welfare.index:
    lat = welfare.loc[i,'lat']
    lng = welfare.loc[i,'lon']
    
    #추출한 정보를 지도에 표시
    icon = folium.Icon(color='green', icon='leaf')
    #tooltip=charger_clean.loc[i,'소재지도로명주소']
    marker = folium.Marker([lat,lng],icon=icon)
    marker.add_to(g_map)

display(g_map)
```

```python
pd.merge?
```

```python
disable_charger=pd.merge(left=disabled_seoul20,right=charger_city,how='left',left_on='통계시군구명',right_on='시군구명').fillna(0)
disable_charger=disable_charger[['통계시군구명','등록장애인수','휠체어충전소개수']]
disable_charger['등록장애인수']=disable_charger['등록장애인수']/100
disable_charger
```

```python
pd.DataFrame.rename?
```

```python
left=disable_charger[['통계시군구명','등록장애인수']]
left['구분']='등록장애인수'
left=left.rename(columns={'등록장애인수':'수'})
right=disable_charger[['통계시군구명','휠체어충전소개수']]
right['구분']='휠체어충전소개수'
right=right.rename(columns={'휠체어충전소개수':'수'})
disable_charger=pd.concat([left,right])
```

```python
charger_ratio=pd.merge(left=disabled_seoul20,right=charger_city,how='left',left_on='통계시군구명',right_on='시군구명').fillna(0)
charger_ratio=charger_ratio[['통계시군구명','등록장애인수','휠체어충전소개수']]
charger_ratio['비율']=charger_ratio['휠체어충전소개수']*100/charger_ratio['등록장애인수']
charger_ratio=charger_ratio.sort_values(by='비율',ascending=False)
fig = plt.figure(figsize=(15, 5))
# bar그래프 그리기
sns.barplot(x='통계시군구명', y='비율',data=charger_ratio)
```
