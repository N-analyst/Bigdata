## Datetime

```python
date_period=pd.date_range(start='20200101',end='20200930',freq='D')
```

## 데이터 받아와서 이름 바꾸기

```python
driver = webdriver.Chrome('C:/Temp/chromedriver')
url="http://m.calltaxi.sisul.or.kr/api/open/newEXCEL0002.asp?key=a4c004f99667232681a84b322cd5a5e9&sDate="
dir_url='C:/Users/won/Downloads/'

# strftime으로 원하는 형식의 str으로 변환.
for date in date_period:
    driver.get(url+date.strftime('%Y%m%d'))
    time.sleep(2)
    files=os.listdir(dir_url)
    idx=files.index('open_data_02_2020-10-25.xls')
    rename(dir_url+files[idx],dir_url+'top100_'+date.strftime('%Y%m%d')+'.xls')
```

## xls파일 합쳐서 xlsx파일 만들기

```python
dir_url="C:/Users/won/Desktop/public_data/top100/"
files=os.listdir(dir_url)
df=pd.DataFrame()

for file in files:
    tables = pd.read_html(dir_url+file,header=0,encoding='cp949')
    df=pd.concat([df,tables[0]],axis=0)
df.to_excel("C:/Users/won/Desktop/public_data/total_top100.xlsx",index=False)
```

```python
charger=pd.read_csv('C:/Users/won/Desktop/public_data/charger_data.csv')
display(charger.head())
```

```python
charger_clean=charger[charger['시도명'].isin(['서울특별시','경기도'])]
charger_clean=charger_clean[charger_clean['시군구명'].isin(city_list)]
charger_clean.head()
```

```python
import folium
map_osm = folium.Map(location=[37.566647,127.978426], zoom_start=8)
map_osm
```

```python
charger_clean=charger_clean.dropna(subset=['위도','경도'])

for i in charger_clean.index:
    lat = charger_clean.loc[i,'위도']
    lng = charger_clean.loc[i,'경도']
    
    #추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng])
    marker.add_to(map_osm)
    
map_osm
```
